```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from wordcloud import WordCloud
import folium
from PIL import Image
from folium import plugins
mpl.style.use(['ggplot'])
```

```python
#membaca data csv menggunakan Pandas
df1 = pd.read_csv('London_Crime1.csv')
df2 = pd.read_csv('London_Crime2.csv')
```

# df_year

```python
df_dummy = df1.head(1000)
```

```python
df_dummy= df_dummy[['borough','year']].value_counts() #mendapatkan nilai dari wilayah dan tahun
df_dummy = df_dummy.to_frame() #menjadikan dataframe
```

```python
df_dummy = pd.DataFrame(df_dummy).reset_index() 
df_dummy.columns = ['borough', 'year','values'] #mengganti nama kolom
```

```python
df_2008 = df_dummy[df_dummy['year']==2008]
df_2008 = df_2008[['borough','values']]
df_2008 = df_2008.set_index('borough')
df_2008.index.name = None
df_2008.rename(columns={'values':2008},inplace = True)
```

```python
df_2009 = df_dummy[df_dummy['year']==2009]
df_2009 = df_2009[['borough','values']]
df_2009 = df_2009.set_index('borough')
df_2009.index.name = None
df_2009.rename(columns={'values':2009},inplace = True)
```

```python
df_2010 = df_dummy[df_dummy['year']==2010]
df_2010 = df_2010[['borough','values']]
df_2010 = df_2010.set_index('borough')
df_2010.index.name = None
df_2010.rename(columns={'values':2010},inplace = True)
```

```python
df_2011 = df_dummy[df_dummy['year']==2011]
df_2011 = df_2011[['borough','values']]
df_2011 = df_2011.set_index('borough')
df_2011.index.name = None
df_2011.rename(columns={'values':2011},inplace = True)
```

```python
df_2012 = df_dummy[df_dummy['year']==2012]
df_2012 = df_2012[['borough','values']]
df_2012 = df_2012.set_index('borough')
df_2012.index.name = None
df_2012.rename(columns={'values':2012},inplace = True)
```

```python
df_2013 = df_dummy[df_dummy['year']==2013]
df_2013 = df_2013[['borough','values']]
df_2013 = df_2013.set_index('borough')
df_2013.index.name = None
df_2013.rename(columns={'values':2013},inplace = True)
```

```python
df_2014 = df_dummy[df_dummy['year']==2014]
df_2014 = df_2014[['borough','values']]
df_2014 = df_2014.set_index('borough')
df_2014.index.name = None
df_2014.rename(columns={'values':2014},inplace = True)
```

```python
df_2015 = df_dummy[df_dummy['year']==2015]
df_2015 = df_2015[['borough','values']]
df_2015 = df_2015.set_index('borough')
df_2015.index.name = None
df_2015.rename(columns={'values':2015},inplace = True)
```

```python
df_2016 = df_dummy[df_dummy['year']==2016]
df_2016 = df_2016[['borough','values']]
df_2016 = df_2016.set_index('borough')
df_2016.index.name = None
df_2016.rename(columns={'values':2016},inplace = True)
```

```python
#Merging DataFrame
df_year = pd.merge(df_2008,df_2009,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2010,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2011,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2012,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2013,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2014,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2015,how='left',right_index=True,left_index=True)
df_year = pd.merge(df_year,df_2016,how='left',right_index=True,left_index=True)
```

```python
df_year.dropna(inplace=True) #drop null
```

# Pie Chart

```python
df_crimetype_top5=df_crimetype.head()
colors_list = ['pink', 'cyan', 'lightcoral', 'lightskyblue', 'silver']
explode_list = [0.1, 0, 0, 0, 0]

df_crimetype_top5['Values'].plot(kind='pie',
                    figsize=(15, 6),
                    autopct='%1.1f%%',   # format persentase
                    startangle=90,       # Angle dari pie chart
                    shadow=True,         # Membuat bayangan
                    labels=None,         # mematikan atau menghilangkan label
                    pctdistance=1.11,    # rasio text dengan pie chart 
                    colors=colors_list,  # menambahkan warna
                    explode=explode_list # menambahkan explode untuk index pertama
                    )

plt.title('Jenis Kejahatan di London') # Menambahkan title

plt.axis('equal')

plt.legend(labels=df_crimetype.index, loc='upper left') # Menambahkan Legend dan lokasinya berasa di atas kiri

plt.show()
```

# Scatter Plot

```python
df_year.columns = list(map(str, df_year.columns))#mengubah kolom menjadi string
years = list(map(str, range(2008, 2017))) #menampung tahun 2008 - 2016 ke variabel years dan berupa string
```

# Word Clauds

```python
word_string = '' #variabel untuk menampung kata
for crime in df2['Crime type']: #looping untuk menampung semua kata
    word_string = word_string + crime + ' '
word_string
```

```python
#Membuat Word Claud 
wordcloud = WordCloud(background_color='white').generate(word_string)

#Menampilkan Word Claud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Folium Maps

```python
df_Maps = df2[['Crime type','Longitude','Latitude']] #menyimpan data ke df_maps
df_Maps = df_Maps.dropna() #drop null
df_Maps.head()
```

```python
latitude = 51.509865
longitude = -0.118092

london_map = folium.Map(location = [latitude, longitude], zoom_start = 12) #membuat map London

crime = plugins.MarkerCluster().add_to(london_map)

#loop untuk mark cluster
for lat, lng, label, in zip(df_Maps.Latitude, df_Maps.Longitude, df_Maps['Crime type']):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(crime)

london_map #menampilkan map London
```
